```python
import pandas as pd
```

```python
fruit_and_fruit_juices_url = "https://raw.githubusercontent.com/yildiramdsa/nutrient_composition_of_common_foods_in_canada_analyzing_the_canadian_nutrient_file/main/data/raw/nutrient_value_fruits_and_fruit_juices_2008_valeur_nutritive_fruits_et_jus_de_fruits-eng.csv"

try:
    fruit_and_fruit_juices_df = pd.read_csv(fruit_and_fruit_juices_url, encoding="utf-8")
except UnicodeDecodeError:
    fruit_and_fruit_juices_df = pd.read_csv(fruit_and_fruit_juices_url, encoding="latin1")

fruit_and_fruit_juices_df.head()
```
